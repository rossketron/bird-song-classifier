```python
import os
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd

audio_fpath = "../data/train_audio/afrsil1/XC125458.ogg"
# audio_fpath = "../data/train_audio/"
# species_list = os.listdir(audio_fpath)
# species_files = []
# for species in species_list:
#     species_files.append(os.listdir(audio_fpath + species))
```

```python
# for species in species_list:
#     for sp_file in species_files:
#         # load image and create spectrogram
#         # https://towardsdatascience.com/audio-deep-learning-made-simple-sound-classification-step-by-step-cebc936bbe5

x, sr = librosa.load(audio_fpath, sr=44100)
print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```
